```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By


# Find the table element using a unique identifier
table=driver.find_element(By.XPATH,'.//*[@id="simple-table-with-pagination"]')

# Initialize an empty dictionary to store the column data
data = {}

# Iterate over the columns of the table
columns = table.find_elements(By.TAG_NAME, 'th')
for column_index, column in enumerate(columns):
    column_name = column.text.strip()
    data[column_name] = []

    # Iterate over the rows of the table and extract the column data
    rows = table.find_elements(By.TAG_NAME, 'tr')
    for row_index, row in enumerate(rows):
        if row_index == 0:
            continue  # Skip the header row
        columns = row.find_elements(By.TAG_NAME, 'td')
        column_data = columns[column_index].text.strip()
        data[column_name].append(column_data)

# Create a pandas DataFrame from the extracted data
#df = pd.DataFrame(data)

# Close the WebDriver
#driver.quit()

# Print the DataFrame
#print(df)
```

```python
sl_no = data['SL.No']
```

```python
state = data['State']
```

```python
Name = data['Name']
```

```python
Address = data['Address']
Email = data['Email']
clas=data['Class']
category = data['Category']
```

```python
df = pd.DataFrame(list(zip(sl_no,state,Name,Address,Email,clas,category)),columns=('sl_no','state','Name','Address','Email','Class','category'))
```
