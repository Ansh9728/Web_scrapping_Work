### Cleaning contact1

```python
df['Contact1'].sample(4)
```

```python
df['Contact1']=df['Contact1'].astype(str)
```

```python
def remove_first_two_digit(number):
    if len(number)>10:
        return number[2:]
    else:
        return number
    
```

```python
df['ContactNum1'] = df['Contact1'].apply(remove_first_two_digit)
```

```python
df['ContactNum1'].head()
```
